###Using Pandas

From the data dictionary on the NYC Open Data portal, we see that “tree_dbh” is a measure of the trunk diameter at 54” (or 137 cm) above the ground. 

**Note** that although the data dictionary gives the metric value for the height, the actual data appear to be solely measured in inches. 

**Remember** that our data is a non-randomly sampled subset of the much larger dataset. As a result, we have to be careful with things like mean and standard deviation. These values could be wildly different with the inclusion of the rest of the data. What we can say is that for the sample of 1,914 trees in our data set, these are the mean and standard deviation numbers. 

The same holds true for the percentiles. That said, the data in our set show that the 50th percentile (or the number under which 50% of the data points fall) is 10 inches. 

**Note** that this is smaller than the mean, which indicates that the data are skewed. In this case, they are “right skewed” meaning that there is a “long tail” on the right of the graph. 

In a “normal” distribution, the mean, median, and mode (and hence the 50th percentile point) are all the same and fall in the middle of the data. The data above and below that point are evenly distributed and form a “bell curve”. 

In a skewed dataset, like the one we have here, the data are not evenly distributed around the center. Our tree data is more heavily clustered among the smaller trunk diameters. The presence of much larger trunks (e.g. 44”) create ‘outliers’ that pull the “tail” of the data toward the right. 

We can get a sense of just how far away from the other numbers the maximum is by looking at the 75th percentile. In our data, 75% of the tree diameters fall below 16”. As we can see, 44” is a good distance from the majority of the measurements in the series. Pandas allows us to measure the amount of skew we have.

````python
tree_diam.skew()
````

**Be aware** that skewed data does not automatically mean “bad” data. In natural systems (like the trunks of trees) normal variation can lead to skewing. The question then becomes what are potential reasons for the skew? It could be our non-random sampling. It could be that some tree species grow better or worse in the urban environment of NYC. It might also be that there are more tree species in our sample that are prone to being smaller–the choice to plant them along with bigger species could also drive the variation. It could be a combination of one or more of these factors, or something else entirely. In other words, we require more context to understand our data. This is critical – **data do not “mean” anything by themselves**. We have to put them into context before the numbers mean anything useful at all.

In addition to isolating entire rows, Pandas also allows us to group rows by the values in their columns. For example, what if we wanted to know which trees were in each region of NYC? We could use the .groupby() method to organize them by “zip_city” (i.e. Richmond Hill, Flushing, etc.) and the common name of the tree.

````python
trees_grouped = trees_df.groupby('zip_city')['spc_common']

for zip_code_pattern in trees_grouped:
    print(zip_code_pattern)
````

**Be aware** that .groupby() is a method associated with series data – sometimes when working with dataframes, you get a series object as a result. The results of our grouping can be a bit hard to read, so let’s isolate things a bit. Let’s find a way to get the information in a form where we have the zip city followed by a list of the trees associated with it.

````python
for item in trees_grouped:
    region = item[0]
    trees = item[1].tolist()
    print(f'{region}:{trees}')
````

**Note** using the .tolist() method allows us to output all of the trees in a neighborhood as a list. If we do not use this method, Pandas will print a line for each entry. 

````python
#get unique species rather than repeats
for item in trees_grouped:
    region = item[0]
    trees = item[1].unique().tolist()
    print(f'{region}:{trees}')
````

````python
#get counts for species by neighborhood
for item in trees_grouped:
    region = item[0]
    trees = item[1].unique().tolist()
    counts = item[1].value_counts().tolist()
    print(f'{region}:{trees}:{counts}')
````

````python
#use .zip() to associate the parallel lists from our analysis into tuples
for item in trees_grouped:
    region = item[0]
    trees = item[1].unique().tolist()
    counts = item[1].value_counts().tolist()
    trees_and_counts = list(zip(trees, counts))

    print(f'{region}:{[tree for tree in trees_and_counts]}')
````

**Question:** What if we wanted to save the data we’ve been working with–what might be a good structure to use and what file format would we employ?


````python
# initialize an empty dictionary for data
trees_dict = {}

#iterate over the dataframe and extract specific information
for item in trees_grouped:
    # region is the neighborhood where the trees are (e.g. Astoria)
    region = item[0]

    # create a unique list of trees to avoid repetitions in species name
    trees = item[1].unique().tolist()

    # count the instances of each species and convert the output to a list
    counts = item[1].value_counts().tolist()
   
    # populate the empty dictionary by setting the first key as region and then creating a dictionary by zipping trees and counts
    trees_dict[region]=dict(zip(trees,counts))
````

**Notice** that we use the zip function we employed in the previous version of this code but now are wrapping the output in a dictionary by using dict() and assigning that to the key “region”.

````python
#write the information as a .json() file
import json

trees_path = 'Tree_Species_and_Counts_by_Location.json'

with open(trees_path, 'w') as json_object:
    json.dump(trees_dict, json_object, indent=2)
    print("file written")
````

````python
def check_for_list(some_stuff):
  temp_list = []
  for item in some_stuff:
    if isinstance(item, list):
      for thing in item:
        temp_list.append(thing)
    else:
      temp_list.append(item)
  return temp_list
````

````python
new_list =check_for_list(list_o_things)
for item in new_list:
  print(item)
````
