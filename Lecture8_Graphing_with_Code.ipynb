##Plotting Data

Thus far we have been working with text or tabular data. This has some real advantages (density of information, organization of information, etc.) but it can sometimes be hard to perceive all of the complexity contained in it. It can be very helpful to put data into a visual format; doing so can help make complex relationships easier to perceive. Reading a bar graph or a scatter plot, for example, may help you, or others viewing the plots, see the shape of the data more quickly than might be possible looking at rows and columns of numbers. 

One of the most used visualization tools in Python is Matplotlib. Its large user base helps ensure that there is a community of people maintaining the code, writing new modules for it, and finding new and innovative ways to use it. Matplotlib provides an array of tools that you can use to output your data visually. The basic format for doing this is to arrange and tidy your data, isolate the dimensions of your data that you want to visualize, and then create a plot object that outputs your data. Matplotlib allows you to work with different types of plots (bar charts, line graphs, scatter plots, pie charts, etc.) so you can choose the format that best fits your data. The library provides mechanisms for creating titles, labeling axes, labeling ticks, creating legends, and other aesthetic/communication elements useful for helping a viewer understand the output. These elements are important because an unlabelled graph removes much-needed context that a viewer needs to make sense of what you, and the data, are trying to communicate. 

**Remember** that data in isolation tell us nothing. We make sense out of data by analyzing, arranging, and creating narratives about what we perceive.



````python
#Introducing plotting -- using numeric relationships for f(x) = x^2 for 1-30
from matplotlib import pyplot as plt
base_numbers = range(1,31)
#use list comprehension to carry out exponentiation on each number during iteration
numbers_squared = [(number**2) for number in base_numbers]

plt.plot(base_numbers, numbers_squared)
plt.show()
````

````python
#modify previous plot to add useful information to make the information more legible
plt.plot(base_numbers, numbers_squared)
plt.title("Squared Values of Integers 1-30")
plt.xlabel("Base Number")
plt.ylabel("Number Squared")
plt.show()
````

````python
#Add graph of f(x) = x^3 to previous graph to show growth rate comparison
base_numbers = range(1,31)
numbers_squared = [(number**2) for number in base_numbers]
numbers_cubed = [(number**3) for number in base_numbers]

plt.plot(base_numbers, numbers_squared)
plt.plot(base_numbers, numbers_cubed)
plt.title("Exponential Growth of Integers 1-30")
plt.xlabel("Base Number")
plt.ylabel("Exponential Outcome")
plt.show()
````

Because our x values are the same for both our y value lists, they will neatly map when we output to the plot. 

**Be aware** that if your x values are not the same for your differing y values, this approach will not work.

**Notice** that matplotlib has given each its unique color. We can control this if we like by adding an argument to each specifying which color we want. For example, we could use ```color = purple``` or even designate the hex value ```color = #944dff```, after we provide the x and y values in our ```plt.plot()``` statement

>```plt.plot(base_numbers, numbers_squared, color=purple))``` 

For now, well stay with the default colors but feel free to play around with them on your own.

````python
#modify previous plot with an added legend for greater legibility
plt.plot(base_numbers, numbers_squared, label='Squared')
plt.plot(base_numbers, numbers_cubed, label='Cubed')
plt.title("Exponential Growth of Integers 1-30")
plt.xlabel("Base Number")
plt.ylabel("Exponential Outcome")
plt.legend()
plt.show()
````

````python
#return to tree data for plotting
import pandas as pd

file_path = '/content/2015_Street_Tree_Census_Tree_Data_First_2K_cleaned.csv'
trees_df = pd.read_csv(file_path)
trees_df.head()
````

````python
#revisit groupby code to get tree species and counts by city region
trees_grouped = trees_df.groupby('zip_city')['spc_common']
trees_counts_all = []
regions_all = []
for item in trees_grouped:
   region = item[0]
   regions_all.append(region)
   trees = item[1].unique().tolist()
   counts = item[1].value_counts().tolist()
   trees_and_counts = list(zip(trees, counts))
   trees_counts_all.append(trees_and_counts)

   print(f'{region}:{[tree for tree in trees_and_counts]}')
````

````python
#see frequency of each tree species in relation to all the others
new_data = trees_df.spc_common.value_counts()
new_data.hist()
````

The output for this gives a shape based on how the data is distributed but we dont get much else from it. We can see that the majority of species have counts between 0 and 50; higher count totals than that are less frequent. While there are plenty of uses for histograms, lets see if we can use a bar graph to give us some better insights. A bar graph is good for plotting discrete categories.

````python
#make bar graph for better insights
new_data = trees_df.spc_common.value_counts()
#initialize empty lists for plotting
species_list = []
counts_list = []
#iterate over data and put data into parallel lists
for item in new_data.iteritems():
    species = item[0]
    species_list.append(species)
    count = item[1]
    counts_list.append(count)

#create and style a plot for output
plt.style.use('seaborn')
plt.bar(species_list, counts_list)
plt.title("Tree Species in NYC by Count")
plt.ylabel("Tree Count")
plt.xticks(rotation=90, size=8)
plt.tight_layout()
plt.show()
````

````python
#make the bars a bit easier to read as individuals and use consistent color
plt.style.use('seaborn')
plt.bar(species_list, counts_list,  color='#17ab08', width=0.5)
plt.title("Tree Species in NYC by Count")
plt.ylabel("Tree Count")
plt.xticks(rotation=90, size=8)
plt.tight_layout()
plt.show()
````

**Remember** that keeping in mind consistency of presentation can be a way of helping your viewer make sense of data that is related. Be aware, however, that you do not want your audience to confuse one set (or subset) of data with another. 

**Lesson:** be attentive to the role that shape and color play in how your audience perceives the information you are conveying.

````python
#return to Callery Pear data
# isolate Callery pear trees in data
just_pears = trees_df.loc[trees_df['spc_common'] == 'Callery pear']

#group pears trees by neighborhood and trunk diameter
pears_grouped = just_pears.groupby('zip_city')['tree_dbh']

neighborhood_list = []
tree_diam_list = []

#iterate over data and isolate specific neighborhoods and trunk data
for item in pears_grouped:
    neighborhood = item[0]
    neighborhood_list.append(neighborhood)
    #use .values to help isolate the items for each neighborhood in the lists
    tree_diam_list.append(item[1].values)
````

**Notice** that we call .values as we iterate to get the list of values associated with each neighborhood. If we tried to plot these lists against the list of neighborhoods, we would get an error because the length of each array (our x values and the various y values we now have) do not match. 

If we distill all of the measurements in each list, which is indexable to a particular neighborhood, to a single measurement, we can plot them. Using **mean** is a good choice. 

````python
#calculate mean diameter from data
from statistics import mean
mean_dia_list = [mean(diam) for diam in tree_diam_list]
````

````python
#plot mean trunk widths per neighborhood
plt.style.use('seaborn')
plt.bar(neighborhood_list, mean_dia_list, color='#17ab08')
plt.title("Mean Trunk Diameter for Callery Pear Trees per NYC Neighborhood")
plt.xlabel("NYC Neighborhood")
plt.ylabel("Average Trunk Diameter (in inches)")
plt.xticks(rotation=90)
plt.show()
````

Because we have a numpy ndarray, we will need to use the .nan_to_num() method to convert our nan values to zeros. To do that, we will need to import numpy.

````python
#We will need numpy to convert the contents of the np_array object
import numpy as np

tlc_counts = np.nan_to_num(tree_location_count_df.values)

tlc_counts
````

````python
#get the mean for all of the counts in our records
count_means = [mean(entry) for entry in tlc_counts]

count_means[:10]
````

Now that we have our y data, lets graph the mean tree count per tree species (we isolate that by calling .index on our data frame).

````python
#plot mean tree count
plt.style.use('seaborn')
plt.bar(tree_location_count_df.index, count_means)
plt.title("Mean Tree Count per Species")
plt.xlabel("Species")
plt.ylabel("Mean Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

**Question** What have we done _wrong_ here?

**Hint** what might the effect of adding so many zeros to our data (when we converted from _nan_) have had on our mean calculation?

````python
#reestablish the values of our counts from our data
tlc_counts = tree_location_count_df.values
````

````python
#use .nanmeans() method to ignore nan values without skewing the mean
count_means_corrected = [np.nanmean(entry) for entry in tlc_counts]

count_means_corrected[:10]
````

````python
#replot mean tree count
plt.style.use('seaborn')
plt.bar(tree_location_count_df.index, count_means_corrected)
plt.title("Mean Tree Count per Species")
plt.xlabel("Species")
plt.ylabel("Mean Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

**Remember** The mean is calculated by summing all values in a series of numbers and then dividing that sum by the total number of entries:

$$
(\frac{1}{n} \sum_{i=i}^{n} x_{i})
$$

While a large number of zeros will not effect the summation portion of the equation, they will effect the mean by driving up the total number of entries. The larger the denominator, the smaller the outcome of division. 

**Lesson:** be sure to really think through any calculations or transformations you want to to use in relation to your data. Be sure that they are appropriate and that you are doing them correctly. Remember: computers are not smart, they will do exactly what you tell them to do. If you tell them to do the incorrect thing, they will as long as it is syntactically correct in relation to the interpreter.

````python
#OPTIONAL: Complex visualization using color mapping

# color mapping using technique from https://stackoverflow.com/questions/26139423/plot-different-color-for-different-categorical-levels-using-matplotlib
health_status = trees_df.health.unique()
health_color = ('blue', 'green', 'orange')
color_map = dict(zip(health_status, health_color))

#scatter plot of species by diameter colored by health (fair, good, poor)
#use advanced handles: https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html
#position plot using: https://matplotlib.org/3.1.1/gallery/userdemo/simple_legend01.html#sphx-glr-gallery-userdemo-simple-legend01-py
import matplotlib.patches as mpatches
plt.scatter(trees_df.spc_common, trees_df.tree_dbh, c=trees_df['health'].map(color_map), marker='p')
plt.xticks(rotation=90, size=7)

#create handlers
fair = mpatches.Patch(color='blue', label='Fair Health')
good = mpatches.Patch(color='green', label='Good Health')
poor = mpatches.Patch(color='orange', label='Poor Health')

#position plot
plt.legend(handles=[fair, good, poor], bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title("NYC Tree Species by Diameter and Health Status")
plt.ylabel("Diameter in Inches")
plt.tight_layout()
plt.show()
````
